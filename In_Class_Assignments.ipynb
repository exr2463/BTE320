```python
#assignment 1 - 1/23/26

sales = float(input('Enter sales ($): '))
tax = float(input('Enter tax rate (%): '))/100
cogs = sales*0.4
grossProfit = sales*0.6
taxes = grossProfit*tax
netIncome = grossProfit-taxes

print(f'Sales: ${sales}')
print(f'Cost of Goods Sold: ${cogs}')
print(f'Gross Profit: ${grossProfit}')
print(f'Taxes: ${taxes}')
print(f'Net Income: ${netIncome}')
```

```python
#assignment 2 - 1/30/26

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, //, **, %): ")

if(op == '+'):
  print(num1 + num2)
elif(op == '-'):
  print(num1 - num2)
elif(op == '*'):
  print(num1 * num2)
elif(op == '/'):
  print(num1 / num2)
elif(op == '//'):
  print(num1 // num2)
elif(op == '**'):
  print(num1 ** num2)
elif(op == '%'):
    print(num1 % num2)
else:
  print("Invalid operator")
```
