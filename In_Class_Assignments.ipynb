```python
#assignment 13

def roi(cost: float, revenue: float) -> float:
  assert cost > 0, "Cost must be positive"
  assert isinstance(cost, (int, float)), "Revenue must be numerical"

  return (revenue - cost) / cost

if __name__ == "__main__":
  c = float(input("enter cost: "))
  r = float(input("enter revenue: "))
  print(roi(c,r))
```

```python
#assignment 14

def loan_payment(principal: float, annual_rate: float, months: int) -> float:
  assert isinstance(principal, (int, float)), "principal must be numerical"
  assert isinstance(annual_rate, (int, float)), "annual_rate must be numerical"
  assert isinstance(months, int), "months must be an integer"

  if principal <= 0:
    raise ValueError("principal must be greater than 0")
  if annual_rate <= 0:
    raise ValueError("annual_rate cannot be negative")
  if months <= 0:
    raise ValueError("months must be greater than 0")

  r = annual_rate / 12.0
  n = months
  P = float(principal)

  if r == 0.0:
    return P / n
  else:
    return r * P / (1.0 - (1.0 + r) ** (-n))

if __name__ == "__main__":
  try:
    P = float(input("enter principal: "))
    apr = float(input("enter annual rate: "))
    n = int(input("enter months: "))
    payment = loan_payment(P, apr, n)
    print(f"monthly payment: ${payment:.2f}")
  except Exception as e:
    print(e)
```

```python
def getValues(n: int) -> list:
  items = []

  for i in range(n):
    try:
      item = float(input("enter a weight value: "))
    except Exception as e:
      item = float("nan")

    items.append(item)
  print(items)
  return items

def calcAverage(alist: list) -> float:
  try:
    return sum(alist[2:]) / len(alist[2:])
  except Exception as e:
    print(e)
    return float("nan")

def main():
  n = int(input("how many weights? "))
  if n < 0:
    raise ValueError("number of items must be positive")

  L = getValues(n)
  avg = calcAverage(L)
  print(f"Average weight: {avg} lbs")

main()
```

```python
#assignment 16

class Names:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal):
    if formal == True:
      print(self.last, ", ", self.first, sep = "")
    else:
      print(self.first)

  def __str__(self):
    return self.first + " " + self.last

myName = Names("Liza","Redmond")
myName.display(True)
myName.display(False)
print(myName)
```

```python
#assignment 17

class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []

  def getName(self):
    return self.__name

  def getCourses(self):
    return self.__courses

  def setName(self, newName):
    assert isinstance(newName, str)
    self.__name = newName

  def __len__(self):
    return len(self.__courses)

  def enroll(self, newCourse):
    if newCourse not in self.__courses:
      self.__courses.append(newCourse)
    else:
      print(self.__name, "is already enrolled in", newCourse)

s = Student("Liza",12345)
print(s.getName())
s.setName("Liza Redmond")
print(s.getName())
print(s.getCourses())
print(s.__len__())
s.enroll("english")
s.enroll("math")
s.enroll("history")
s.enroll("spanish")
print(s.getCourses())
s.enroll("english")
print(s.getCourses())
print(s.__len__())
```

```python
#assignment 18

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def distance(self, other):
    dx = (self.x - other.x)**2
    dy = (self.y - other.y)**2
    return (dx+dy)**0.5

class Rectangle:
  def __init__(self, point1, point2, point3, point4):
    self.point1 = point1
    self.point2 = point2
    self.point3 = point3
    self.point4 = point4
    if (self.point1.x == self.point2.x):
      self.width = abs(self.point1.x - self.point3.x)
      self.height = abs(self.point1.y - self.point2.y)
    else:
      self.width = abs(self.point1.x - self.point2.x)
      self.height = abs(self.point1.x - self.point3.x)

  def area(self):
    return self.width * self.height

p1 = Point(0,0)
p2 = Point(0,4)
p3 = Point(4,4)
p4 = Point(4,0)
r = Rectangle(p1,p2,p3,p4)
print(r.area())
```

```python
#assignment #19
class Employee:
  def __init__(self, name, hourly_wage, hours):
    self.name = name
    self.__hourly_wage = hourly_wage
    self.hours = hours

  def display_info(self):
    print(f"Employee: {self.name}\nHourly wages: {self.__hourly_wage}")

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self, new_wage):
    if new_wage <= 0:
      raise ValueError("wage cannot be negative")
    self.__hourly_wage = new_wage

  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def __str__(self):
    return f"Employee: {self.name}\nHourly wage: ${self.__hourly_wage}/hour"

class Manager(Employee):
  def __init__(self, name, wage, hours, bonus):
    super().__init__(name, wage, hours)
    self.bonus = bonus

  def calculate_pay(self):
    return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly payment is ${employee.calculate_pay()}")
print(f"Manager {manager.name} weekly payment is ${manager.calculate_pay()} that includes a {manager.bonus * 100}% bonus")
```
