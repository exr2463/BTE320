```python
#assignment 1 - 1/23/26

sales = float(input('Enter sales ($): '))
tax = float(input('Enter tax rate (%): '))/100
cogs = sales*0.4
grossProfit = sales*0.6
taxes = grossProfit*tax
netIncome = grossProfit-taxes

print(f'Sales: ${sales}')
print(f'Cost of Goods Sold: ${cogs}')
print(f'Gross Profit: ${grossProfit}')
print(f'Taxes: ${taxes}')
print(f'Net Income: ${netIncome}')
```

```python
#assignment 2 - 1/30/26

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, //, **, %): ")

if(op == '+'):
  print(num1 + num2)
elif(op == '-'):
  print(num1 - num2)
elif(op == '*'):
  print(num1 * num2)
elif(op == '/'):
  print(num1 / num2)
elif(op == '//'):
  print(num1 // num2)
elif(op == '**'):
  print(num1 ** num2)
elif(op == '%'):
    print(num1 % num2)
else:
  print("Invalid operator")
```

```python
#assignment 5 - 2/11/26
import random
again = input("would you like to play the game? (y/n): ").lower()

while again == 'y':
  userMove = input("enter \"rock\",\"paper\", or \"scissors\" here: ").lower()
  computerMove = random.choice(['rock','paper','scissors'])

  print(f"user's move: {userMove}")
  print(f"computer's move: {computerMove}")

  if(userMove == computerMove):
    print("it's a tie!")
  elif ((userMove == "rock" and computerMove == "scissors") or (userMove == "paper" and computerMove == "rock") or (userMove == "scissors" and computerMove == "paper")):
    print("user won!")
  else:
    print("computer won!")

  again = input("play again? (y/n): ").lower()


print("thanks for playing!")

```
