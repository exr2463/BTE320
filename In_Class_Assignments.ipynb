```python
#assignment 1

sales = float(input('Enter sales ($): '))
tax = float(input('Enter tax rate (%): '))/100
cogs = sales*0.4
grossProfit = sales*0.6
taxes = grossProfit*tax
netIncome = grossProfit-taxes

print(f'Sales: ${sales}')
print(f'Cost of Goods Sold: ${cogs}')
print(f'Gross Profit: ${grossProfit}')
print(f'Taxes: ${taxes}')
print(f'Net Income: ${netIncome}')
```

```python
#assignment 2

#Inputs: demand, reorder cost, holding cost
demand = float(input("Enter the projected demand (units/year): "))
reorderCost = float(input("Enter the reorder cost ($/order): "))
holdingCost = float(input("Enter the holding cost ($/year/unit): "))

#Process: calculate EOQ
eoq = (2 * demand * reorderCost / holdingCost) ** 0.5

#Output: display EOQ value
print(f"Economic Order Quantity: {eoq}")
```

```python
#assignment 3

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, //, **, %): ")

if(op == '+'):
  print(num1 + num2)
elif(op == '-'):
  print(num1 - num2)
elif(op == '*'):
  print(num1 * num2)
elif(op == '/'):
  print(num1 / num2)
elif(op == '//'):
  print(num1 // num2)
elif(op == '**'):
  print(num1 ** num2)
elif(op == '%'):
    print(num1 % num2)
else:
  print("Invalid operator")
```

```python
#assignment 6
import random
again = input("would you like to play the game? (y/n): ").lower()

while again == 'y':
  userMove = input("enter \"rock\",\"paper\", or \"scissors\" here: ").lower()
  computerMove = random.choice(['rock','paper','scissors'])

  print(f"user's move: {userMove}")
  print(f"computer's move: {computerMove}")

  if(userMove == computerMove):
    print("it's a tie!")
  elif ((userMove == "rock" and computerMove == "scissors") or
    (userMove == "paper" and computerMove == "rock") or
    (userMove == "scissors" and computerMove == "paper")):
    print("user won!")
  else:
    print("computer won!")

  again = input("play again? (y/n): ").lower()


print("thanks for playing!")

```

```python
#assignment 7
userNum = float(input("enter a number: "))
numbers = []

while userNum != 0:
  numbers.append(userNum)
  userNum = float(input("enter a number (0 to quit): "))

numbers.sort(reverse = True)

print("sorted list:",numbers)

print("end program.")
```
